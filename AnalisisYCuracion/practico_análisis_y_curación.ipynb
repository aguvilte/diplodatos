```
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

ds_pesos = pd.read_csv('https://raw.githubusercontent.com/DiploDatos/AnalisisYCuracion/master/input/wtageinf.csv')
```

```
ds_pesos
```

```
mask_ninos = ds_pesos['Sex'] == 1
mask_ninas = ds_pesos['Sex'] == 2

ds_ninos = ds_pesos[mask_ninos]
ds_ninas = ds_pesos[mask_ninas]

percentiles = np.array(ds_pesos.iloc[:,5:].columns)
percentiles
```

```
ds_ninos.plot(x='Agemos', y=percentiles, kind='line')
plt.legend()
```

```
ds_ninas.plot(x='Agemos', y=percentiles, kind='line')
plt.legend()
```

```
ds_provincias = pd.read_json('https://raw.githubusercontent.com/DiploDatos/AnalisisYCuracion/master/input/pasos.json')
```

```
ds_provincias
```

```
provincias = ds_provincias['province'].unique()
provincias
```

```
agrup_prov = ds_provincias.groupby('province').count()
agrup_prov = agrup_prov.iloc[:,:1]
```

```
plt.figure(figsize=(10,6))
sns.countplot(data=agrup_prov, x=ds_provincias.province)
plt.xticks(rotation=90)
```
