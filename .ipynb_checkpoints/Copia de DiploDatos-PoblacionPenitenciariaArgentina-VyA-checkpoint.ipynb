```python
data_raw2 = pd.read_csv('../datasets/sneep_2002_2017_diplodatos.zip',compression='zip',index_col=0)
```

```python
import seaborn
```

```python
plt.figure(figsize=(10,6))
seaborn.lineplot(data=data_raw2, y='edad', x='anio_censo')
plt.title('Evolución de edad promedio de los internos (2002-2017)', size=18)
plt.ylim(25, 40)
plt.ylabel('Edad')
plt.xlabel('Año')
seaborn.despine(left=True)
```

```python
data_male = data_raw2[data_raw2['genero_descripcion'] == 'Masculino']
data_no_male = data_raw2[data_raw2['genero_descripcion'] != 'Masculino']

seaborn.set(rc={'figure.figsize':(10,6)})
seaborn.countplot(
    x='anio_censo', 
    data=data_male,
    hue='genero_descripcion')

plt.title('Cantidad de hombres (2002-2017)', size=20)
seaborn.despine(left=True)
```
